```python
search_item=driver.find_element(By.XPATH,'/html/body/div[3]/div/div[3]/header/div[1]/div[2]/div/div/div[1]/div/div/input')
search_item.send_keys('shoes')
search_item.click()
time.sleep(5)
```

```python
shoes=driver.find_element(By.XPATH,'/html/body/div[3]/div/div[3]/header/div[1]/div[2]/div/div/div[2]/div[2]/div[1]/ul/li[1]/a/span')
shoes.click()
```

```python
# Table.1.1:
shoe_name=[]
category=[]
no_of_colors=[]
price=[]
```

```python
for _ in range(0,1):
    driver.execute_script("window.scrollBy(0,10)")
             
    # to scrap shoe_name
    try:
        name_tag=driver.find_elements(By.XPATH,'//div[@class="product-card__title"][1]')
        for i in name_tag:
            shoe_name.append(i.text)   
            
    except NoSuchElementException:
         shoe_name.append('-')       
    
    # to scrap category:
    try:
        category_tag=driver.find_elements(By.XPATH,'//div[@class="product-card__subtitle"][1]')
        for i in category_tag:
            
            result=re.search(r"\bWomen\b|\bMen\b|\bkid's\b|\bchild\b",str(i.text))
            if result is None:
                category.append(None)
                
            else:    
                category.append(result.group(0))
            
            #category.append(i.text)
            
        
    except NoSuchElementException:
         category.append('-')       
    
    # to scrap no_of_colors:
    
    try:
        colors_tag=driver.find_elements(By.XPATH,'//div[@class="product-card__count-item"][1]')
        for i in colors_tag:
            no_of_colors.append(i.text) 
            
    except NoSuchElementException:
         no_of_colors.append('-')   
            
    # to scrap price
    try:
        price_tag=driver.find_elements(By.XPATH,'//div[@class="product-card__price"][1]')
        for i in price_tag:
            price.append(i.text)
            
    except NoSuchElementException:
         price.append('-') 
            
     #to scrap url:
    shoe_url=[]
    try:
        url_tag=driver.find_elements(By.XPATH,'//*[@data-testid="product-card__link-overlay"]')
        for i in url_tag:
            shoe_url.append(i.get_attribute("href"))
            
    except NoSuchElementException:
         shoe_url.append('-')           
            
                
print(len(shoe_name),len(category),len(no_of_colors),len(price),len(shoe_url))    
```

```python
table1={"SHOE_NAME":shoe_name,"CATEGORY":category,"NO_OF_COLORS":no_of_colors,"PRICE":price}
df1=pd.DataFrame(table1)
df1
```
